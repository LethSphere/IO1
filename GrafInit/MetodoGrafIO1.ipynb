```python
import sympy as sym
import matplotlib.pyplot as pyp
import numpy as num

#Violeth Valmont Azahar - 20181020010
#Esteban Mauricio Lopez Duarte-20181020005
#Julian Alejandro Sarmiento Linares -20181020042
```

```python
def plot_graf(graf, inicio, fin, x, y):
  pyp.imshow((graf).astype(int), extent=(x.min(),x.max(),y.min(),y.max()), origin="lower", cmap="Greens", alpha = 0.4)
  x = num.linspace(inicio, fin, 2000)  
  pyp.xlim(inicio, fin)
  pyp.ylim(inicio, fin)
  pyp.xlabel(r'$x$')
  pyp.ylabel(r'$y$')
  pyp.legend(bbox_to_anchor=(1.09, 1), loc=2, borderaxespad=0.)
```

```python
#Ejercicio 1 2x+y < 5
dis = num.linspace(-20, 20, 500)
x, y = num.meshgrid(dis, dis)
s = num.linspace(-20, 20, 2000)
f = 2 * x + y < 5
y2 = 5 - 2 * s
pyp.plot(s, y2, label='$2x+y<5$')
plot_graf(f, -15, 15, x, y)
```

```python
#Ejercicio 2 y <= 5

dis = num.linspace(-20, 20, 500)
x, y = num.meshgrid(dis, dis)
s = num.linspace(-20, 20, 2000)
y2 = 5 + s * 0
f = y <= 5
pyp.plot(s, y2, label='$y\leq5$')
plot_graf(f, -15, 15, x, y)
```

```python
#Ejercicio 3 2 (2x - y) < 2 (x + y) - 4 

dis = num.linspace(-15, 15, 300)
x, y = num.meshgrid(dis, dis)
s = num.linspace(-15, 15, 2000)
f = 2 * (2 * x - y) < 2 * (x + y) - 4
y2 = s / 2 + 1
pyp.plot(s, y2, label=r'$2 (2x - y) < 2 (x + y) - 4$')

plot_graf(f, -5, 10, x, y)
```

```python
'''Ejercicio 4
      2x+y>3
      2y-1>0
      x >= y
'''
dis = num.linspace(-15, 15, 300)
x, y = num.meshgrid(dis, dis)
s = num.linspace(-15, 15, 2000)
f = (2 * x + y > 3) & (2 * y - 1 > 0) & (x >= y)
y2 = 3 - 2 * s
pyp.plot(s, y2, label='$2x+y>3$')
y3 = 1 / 2 + s * 0
pyp.plot(s, y3, label='$2y-1>0$')
y4 = s
pyp.plot(s, y4, label='$x\geq y$')
plot_graf(f, -15, 15, x, y)
```

```python
'''Ejercicio 5
      2x+3y <= 60 
      x >= 0
      y >= 0
'''
d = num.linspace(-15, 45, 600)
x, y = num.meshgrid(d, d)
s = num.linspace(-15, 45, 2000)
f = (2 * x + 3 * y <= 60) & (x >= 0) & (y >= 0)
y2 = (60 - 2 * s) / 3
pyp.plot(s, y2, label='$2x+3y\leq 60$')
y3 = s * 0
pyp.plot(s, y3, label='$y \geq 0$')
pyp.plot(num.zeros_like(s), s, label='$x\geq 0$')
plot_graf(f, -20, 45, x, y)
```

```python
'''Ejercicio 6
      p = 4x+6y
    sujeto a:
      2x+y<=180
      x+2y<=160
      x >= 0
      x+y<= 100
      y >=0
'''

d = num.linspace(-15, 105, 600)
x, y = num.meshgrid(d, d)
s = num.linspace(-10, 100, 2000)
f = (2 * x + y <= 180) & (x + 2 * y <= 160) & (x >= 0) & (x + y <= 100) & (y >= 0)
y2 = 180 - 2 * s
pyp.plot(s, y2, label='$2x+y\leq 180$')
y3 = (160 - s) / 2
pyp.plot(s, y3, label='$x+2y\leq 160$')
y5 = 100 - s
pyp.plot(s, y5, label='$x+y\leq 100$')
y4 = s * 0
pyp.plot(s, y4, label='$y \geq 0$')
pyp.plot(num.zeros_like(s), s, label=r'$x\geq 0$')
pyp.plot(s, y2, s, y3)
xi, yi = sym.symbols('xi, yi')
points = [(0, 0), (0, 80)]
px, px = [], []
cons = [180 - 2 * xi, (160 - xi) / 2, 100 - xi, xi * 0]
for i in range(len(cons) - 1):
  for k in range(i + 1, len(cons)):
    eq = sym.Eq(cons[i], cons[k])
    sol = sym.solve([eq], (xi, yi))  
    eqst = str(cons[i])
    sol2 = eval(eqst, {'xi': sol[xi]})
    points.append((sol[xi], sol2))   
points.remove((100, 0))
points.remove((200/3, 140/3))
points.remove((160, 0))
p = set(points)
for pair in p:
  xt, yt = pair
  pyp.plot(xt, yt, 'ro')
plot_graf(f, -10, 100, x, y)
max = '4 * x+6 * y'
value = -100000
xm, ym = 0, 0
for pair in p:
  xt, yt = pair
  new_value = eval(max, {'x': xt, 'y': yt})
  if new_value >= value:
    value = new_value
    xm = xt
    ym = yt
print('Valor Max: {}, coordenadas: ({}, {})'.format(value, xm, ym))
```

```python
'''Ejercicio 7
      p = 200x+100y
      sugeto a:
      5x+3y<=105
      2x+4y<=70
      x >= 0
      y >=0
'''

d = num.linspace(-15, 105, 600)
x, y = num.meshgrid(d, d)
s = num.linspace(-10, 40, 2000)
f = (5 * x + 3 * y <= 105) & (2 * x + 4 * y <= 70) & (x >= 0) & (y >= 0)
y2 = (105 - 5 * s) / 3
pyp.plot(s, y2, label=r'$5x+3y\leq 105$')
y3 = (70 - 2 * s) / 4
pyp.plot(s, y3, label=r'$2x+4y\leq 70$')
y4 = s * 0
pyp.plot(s, y4, label=r'$y \geq 0$')
pyp.plot(num.zeros_like(s), s, label=r'$x\geq 0$')
pyp.plot(s, y2, s, y3)
xi, yi = sym.symbols('xi, yi')
points = [(0, 0), (0, 35/2)]
px, px = [], []
cons = [(105 - 5 * xi) / 3, (70 - 2 * xi) / 4, xi * 0]
for i in range(len(cons) - 1):
  for k in range(i + 1, len(cons)):
    eq = sym.Eq(cons[i], cons[k])
    sol = sym.solve([eq], (xi, yi))
    eqst = str(cons[i])
    sol2 = eval(eqst, {'xi': sol[xi]})
    points.append((sol[xi], sol2))
points.remove((35, 0))
p = set(points)
for pair in p:
  xt, yt = pair
  pyp.plot(xt, yt, 'ro')
plot_graf(f, -10, 40, x, y)
max = '200 * x+6 * y'
value = -100000
xm, ym = 0, 0
for pair in p:
  xt, yt = pair
  new_value = eval(max, {'x': xt, 'y': yt})
  if new_value >= value:
    value = new_value
    xm = xt
    ym = yt
print('valor Max: {}, coordenadas: ({}, {})'.format(value, xm, ym))
```
